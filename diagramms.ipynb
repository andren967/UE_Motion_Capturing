```python
# Libraries
import matplotlib.pyplot as plt
import pandas as pd
from math import pi

lM1 = ['P1_M1','P2_M1','P3_M1','P4_M1','P5_M1','P6_M1','P7_M1']
color = iter(cmap(np.linspace(0, 1, 7)))
 
# Set data
df = pd.DataFrame({
'group': ['A','B','C','D','E','F','G'],
'M1': [99.09, 109.48, 92.2,  92.14, 92.25, 98.71, 91.71],
'M2': [106.19, 113.21, 101.62, 83.92, 151.9, 131.33, 84.69],
'M3': [96.33, 102.9, 96.72, 96.2, 99.23, 98.94, 96.21]
})
 
# number of variable
categories=list(df)[1:]
N = len(categories)
 
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
# Initialise the spider plot
ax = plt.subplot(111, polar=True)

# If you want the first axis to be on top:
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
 
# Draw one axe per variable + add labels
plt.xticks(angles[:-1], categories, color='black', size=8)
 
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([80,100,110], ["0.80","1.00","1.10"], color="grey", size=7)
plt.ylim(80,120)
 
# Plot data
#Persons
for x in range(0,7): 
    # We are going to plot the first line of the data frame.
    # But we need to repeat the first value to close the circular graph:
    values=df.loc[x].drop('group').values.flatten().tolist()
    values += values[:1]
    values

    label1=lM1[x]

    c=next(color)

    ax.plot(angles, values, linewidth=1, linestyle='solid', c=c,label=label1, alpha=0.8)
 
    # Fill area
    #ax.fill(angles, values, c=c, alpha=0.1)

#Beat
values=[92.77, 85.13, 96.39]
values += values[:1]

ax.plot(angles, values, linewidth=1, linestyle='solid',color="C0",label="music")

plt.legend(loc="lower right")
plt.title("comparing interbeat and interstep intervall\nafter 3 seconds")
plt.legend(loc="upper left")
plt.savefig("../UE_Motion_Capture/pictures/MAll_comparing_inter.png")

# Show the graph
plt.show()
```

M1 - Boxplots

```python
#read M1 music interbeat intervalls
interbeatM1=[]
with open("../UE_Motion_Capture/InterstepIntervals/M1_InterbeatInterval.txt", "r") as file1:
    f_list = [float(i) for line in file1 for i in line.split(',') if i.strip()]
    interbeatM1 = f_list[0:125]
```

```python
lM1 = ['P1_M1','P2_M1','P3_M1','P4_M1','P5_M1','P6_M1','P7_M1']
dictIntervalls = {}

for string in lM1:
    text_file = open("../UE_Motion_Capture/InterstepIntervals/" + string + "_InterstepInterval.txt", "r")
    lines = text_file.readlines()
    text_file.close()
    lines = [s.replace("\n", '') for s in lines]
    lines = [eval(i)/100 for i in lines]

    dictIntervalls[string]= lines
```

```python
fig, ax = plt.subplots()
ax.boxplot(dictIntervalls.values())
ax.set_xticklabels(dictIntervalls.keys())
plt.hlines(statistics.mean(interbeatM1)/100,0,8,label="music (mean)")

plt.ylim([0.20,0.90])
plt.xticks(np.arange(1,8),labels=['P1','P2','P3','P4','P5','P6','P7'])
plt.title("M1: interstep/interbeat intervalls")
plt.ylabel("interstep/interbeat intervall in s")
plt.legend(loc="upper left")
plt.savefig("../UE_Motion_Capture/pictures/M1_Boxplots.png")
```
